```python
import pandas as pd
import logomaker
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# Příkladová data: frekvence aminokyselin v pozici (AAs vs. pozice)
counts_df = pd.DataFrame({
    'A': [0.4, 0.1],
    'C': [0.1, 0.2],
    'D': [0.2, 0.05],
    'E': [0.1, 0.4],
    'F': [0.2, 0.25],
}, index=[0, 1])

# Patogenita jednotlivých aminokyselin (0 = benigní, 1 = patogenní)
pathogenicity_scores = {
    'A': 0.1,
    'C': 0.8,
    'D': 0.6,
    'E': 0.9,
    'F': 0.3,
}

# Definuj barevné mapování (červená = více patogenní, modrá = méně)
cmap = plt.cm.get_cmap("coolwarm")  # můžeš zvolit i jinou, např. 'viridis', 'plasma'

color_scheme = {
    aa: mcolors.to_hex(cmap(score)) for aa, score in pathogenicity_scores.items()
}

# Vytvoř logo
logo = logomaker.Logo(counts_df,
                      color_scheme=color_scheme,
                      font_name='Arial Rounded MT Bold')

# Titulek a úpravy
logo.ax.set_title('Sekvenční logo s barevným kódováním patogenity')
logo.ax.set_ylabel('Relativní četnost')
logo.ax.set_xlabel('Pozice')

plt.show()
```
